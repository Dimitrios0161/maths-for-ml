### SVD for Image Compression

```python
from matplotlib.image import imread
import matplotlib.pyplot as plt
import os
import numpy as np

plt.rcParams['figure.figsize'] = [16, 8]
```

```python
#print image
A = imread(os.path.join('.','images','psychedelic-pluto_pca.png'))
img_col = plt.imshow(A)
plt.show()
```

```python
# Convert RGB to grayscale - an indepth article about the conversion approach can be found here https://e2eml.school/convert_rgb_to_grayscale.html
B = np.mean(A,2)
img_gray = plt.imshow(B)
img_gray.set_cmap('gray')
plt.axis('off')
plt.show()
```

```python
U, S, VT = np.linalg.svd(B,full_matrices=False)
S = np.diag(S)

j=0
for r in (5,20,100):
    Xapprox = U[:,:r] @ S[0:r,:r] @ VT[:r,:]
    plt.figure(j+1)
    j += 1
    img = plt.imshow(Xapprox)
    img.set_cmap('gray')
    plt.axis('off')
    plt.title('r=' + str(r))
    plt.show()
```
